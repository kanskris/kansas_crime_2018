```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, stdev
from mpl_toolkits.mplot3d import Axes3D


#load data from file to dataframes
file_to_load ="Resources/KCPD_Crime_Data_2018.csv"
raw_data = pd.read_csv(file_to_load)
raw_data.head()
```

```python
totalByArea = raw_data.groupby(['Area'])['Area'].count()
totalByArea
```

```python
Area = ["CPD", "EPD", "MPD", "NPD", "OSPD", "PD1", "SCP", "SPD"]
x_axis = np.arange(len(Area))

Totals = [32427, 34723, 24221, 9200, 664, 155, 11130, 15107]
y_axis = np.arange(len(Totals))
```

```python
x_axis = np.arange(0, len(Area))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.title("Kansas City Crime Division Data")
plt.ylabel('Totals by Area')
plt.xlabel('Divisions')

plt.xlim(-0.75, len(Area)-.25)
plt.ylim(0, max(Totals) + 10000)

plt.bar(x_axis, Totals, facecolor="aqua", alpha=0.75, align="center", edgecolor='blue')
plt.xticks(tick_locations, Area)
plt.show()
```

```python
# data
Area = ["CPD", "EPD", "MPD", "NPD", "OSPD", "PD1", "SCP", "SPD"]
Totals = [32427, 34723, 24221, 9200, 664, 155, 11130, 15107]

# find the mean using the mean() function from the statistics library

Totals_mean = mean(Totals)

# find the standard deviation using the stdev() function from the statistics library

Totals_stdev = stdev(Totals)

# build a bar plot and save it as a .png image
plt.bar(['Totals'],[Totals_mean],yerr=[Totals_stdev], capsize=10)
plt.xlabel('Divisions')
plt.ylabel('Totals by Area')
plt.title('Kansas City Crime Division Data')
plt.savefig('plot.png', dpi = 300)
plt.show()
```
