```python
#import libraries
import pandas as pd
import numpy as np

#load data from file to dataframes
file_to_load ="Resources/KCPD_Crime_Data_2018.csv"
raw_data = pd.read_csv(file_to_load)
raw_data.head()
```

```python
#clean up of sex, race, age
values = {'Sex':"U",'Race':"U",'Age':0,'Area':"U",'Beat':"U"}
clean_data = raw_data.fillna(value=values)
clean_data.head()
```

```python
clean_data.to_csv("Resources/clean_data.csv")
```
